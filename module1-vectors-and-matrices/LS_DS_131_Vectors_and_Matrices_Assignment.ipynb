```
def dot(A,B): 
  #Calculates the matrix product or dot product of A and B where applicable.
  
  #Get the dimensions of A and B
  A_shape = A.shape
  B_shape = B.shape

  #If the dimensions of A and B are equal and they're one dimensional vectors (ie a single row vector), then take the dot product
  if A_shape==B_shape and A_shape[0]==1: 
    result = (A*B).sum()
    return result
  
  #Check dimensions of each to see if it's legal to matrix multiply A and B.
  elif A_shape[1] == B_shape[0]: 
    
    #initialize result matrix
    result = np.zeros((A_shape[0], B_shape[1]))

    #Loop through each row of A and take the dot product of that row with each column of B
    for i in range(A_shape[0]): 
      for j in range(B_shape[1]): 
        dot_product = np.dot(A[i], B[:, j])

        #populate result matrix in location [i,j] with dot_product
        result[i,j] = dot_product

    return result
  
  #If A and B can't be multiplied in any way legally, return error message
  else: 
    print("The dimensions of A and B are incompatible to be multiplied together.")
```
